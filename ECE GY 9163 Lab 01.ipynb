# Step 2 Calculating Infogain and Building Feature Matrices

Calculate the total spam emails and total ham emails:

This is done by opening each folder and checking if the text file contains the string 'spmsg'
if true the counter for spam increases else counter for ham increases.

```
import pandas as pd

def FindTotalSpamHam(train_dir,ts,th):
    emails = [os.path.join(train_dir,f) for f in os.listdir(train_dir)]
    for mail in emails:
      if os.path.basename(mail).find('spmsg'):
        th=th+1
      else:
        ts=ts+1
    return ts,th
                                     
    
    
```

Start Building the dataframe

We first open the directory and go through each email then we use the split function to obtain a list of words which is then put into a dataframe.

```
def make_Dictionary(train_dir):
    emails = [os.path.join(train_dir,f) for f in os.listdir(train_dir)]
    for mail in emails:
      with open(mail) as m:
        data = m.read()
        list = data.split()
        df2 = pd.DataFrame(list,columns=['word'])
    return df2
```

2.3 Calculate the number of times the word aappears in ham or spam messages

```

def spamnhamscores(train_dir,dictionary):
  
  emails = [os.path.join(train_dir,f) for f in os.listdir(train_dir)]
  for mail in emails:
    with open(mail) as m: 
      data = m.read() 
      list = data.split()

      for j in range(len(dictionary)):
        spamcounter=0
        hamcounter=0
        if os.path.basename(m.name).find('spmsg'):
          
          if dictionary.iloc[j,0] in list:
            
            hamcounter=hamcounter+1
        else:
          
          if dictionary.iloc[j,0] in list:
            
            spamcounter=spamcounter+1        
        dictionary.iloc[j,1] = dictionary.iloc[j,1]+hamcounter
        dictionary.iloc[j,2] = dictionary.iloc[j,2]+spamcounter 
  return dictionary
```

Calculate the Information Gain

```
import math
def calculateig(dictionary,totalspam,totalham):
  pden = totalspam+totalham
  p = totalham/pden
  hc = (-1*p*(np.log(p)))-((1-p)*(np.log(1-p))) 
  for j in range(len(dictionary)):
    ham = dictionary.iloc[j,1]
    spam = dictionary.iloc[j,2]
    hamf = totalham-ham
    spamf = totalspam-spam
    naden = (spamf)+(hamf)
    aden = ham+spam
    #hhna = (hamf/pden)*np.log(hamf/naden)
    hsna = (spamf/pden)*np.log(spamf/naden)
    #hha = (ham/pden)*np.log(ham/aden)
    hsa = (spam/pden)*np.log(spam/aden)
    #hcx = (-1*(hhna+hsna+hha+hsa))
    hcx = (-1*(hsa+hsna))
    ig = hc - hcx
    dictionary.iloc[j,3] = ig

  return dictionary
```

Building an empty dataframe for the feature matrix

```
def newdf(train_dir,df):
  array = []
  for c in range(len(df)):
    array.append(df.iloc[c,0])
  emails = [os.path.join(train_dir,f) for f in os.listdir(train_dir)]
  dfcols = len(array)
  dfrows = len(emails)
  dfdata = np.zeros(shape=(dfrows,dfcols))
  newdf = pd.DataFrame(data=dfdata,columns=array)
  return newdf
```

Populating the feature matrices for binary features

```
def populatedata(train_dir,df):
  emails = [os.path.join(train_dir,f) for f in os.listdir(train_dir)]
  counter = 0
  for mail in emails:
    with open(mail) as m:
      data = m.read()
      lst = data.split()
      i=emails.index(m.name)
      for j in df.columns:
        counter=0
        if j in lst:
          counter = 1
        df.iloc[i,df.columns.get_loc(j)] = counter   
  return df
```

Populating Feature Matrix for term frequency

```
import re 
def findtermfreq(train_dir,df):
  emails = [os.path.join(train_dir,f) for f in os.listdir(train_dir)]
  for mail in emails:
    lst = []
    with open(mail) as m:
      data = m.read()
      lst = data.split()
      i=emails.index(m.name)
      for j in df.columns:
        count=lst.count(j) 
        df.iloc[i,df.columns.get_loc(j)] = count   
  return df
```

# Step 3 : Running Bernoulli Classifiers

Building the Naive bayes classifer according to the formula understood

```
def NaiveBayes(df,ig,totalsp,totalh):
  labels=[]
  for i in range(len(df)):
    pspame=1
    phame = 1
    
    for j in df.columns:
      
      if df.iloc[i,df.columns.get_loc(j)]==1:
        for k in range(len(ig)):
          if j==ig.iloc[k,0]:
            pspame=pspame*((ig.iloc[k,2])/totalsp)
            phame = phame*((ig.iloc[k,1])/totalh)
      else:
        for k in range(len(ig)):
          if j==ig.iloc[k,0]:
            pspame=pspame*(1-((ig.iloc[k,2])/totalsp))
            phame = phame*(1-((ig.iloc[k,1])/totalh))
    if pspame>phame:
      labels = np.append(labels,0)
    else:
      labels = np.append(labels,1)  
  return labels
```

Building Multinomial Naive Bayes Classifier according to formula understood

```
def MultinomialNaiveBayes(df,ig,totalsp,totalh):
  labels=[]
  for i in range(len(df)):
    pspame=1
    phame = 1
    for j in df.columns:
      if df.iloc[i,df.columns.get_loc(j)]>=1:
        for k in range(len(ig)):
          if j==ig.iloc[k,0]:
            pspame=df.iloc[i,df.columns.get_loc(j)]*pspame*((ig.iloc[k,2])/totalsp)
            phame = df.iloc[i,df.columns.get_loc(j)]*phame*((ig.iloc[k,1])/totalh)
      else:
        for k in range(len(ig)):
          if j==ig.iloc[k,0]:
            pspame=pspame*(1-((ig.iloc[k,2])/totalsp))
            phame =phame*(1-((ig.iloc[k,1])/totalh))
    if pspame>phame:
      labels = np.append(labels,0)
    else:
      labels = np.append(labels,1)
  return labels
```

# Step 5 : Building an Adversarial Classifier

Recalculating IG in Terms of Ham Emails

```
def calculateig2(dictionary,totalspam,totalham):
  pden = totalspam+totalham
  p = totalham/pden
  hc = (-1*p*(np.log(p)))-((1-p)*(np.log(1-p))) 
  for j in range(len(dictionary)):
    ham = dictionary.iloc[j,1]
    spam = dictionary.iloc[j,2]
    hamf = totalham-ham
    spamf = totalspam-spam
    naden = (spamf)+(hamf)
    aden = ham+spam
    hhna = (hamf/pden)*np.log(hamf/naden)
    #hsna = (spamf/pden)*np.log(spamf/naden)
    hha = (ham/pden)*np.log(ham/aden)
    #hsa = (spam/pden)*np.log(spam/aden)
    #hcx = (-1*(hhna+hsna+hha+hsa))
    hcx = (-1*(hha+hhna))
    ig = hc - hcx
    dictionary.iloc[j,4] = ig

  return dictionary
```

Calculating Cost Of Email

```
def CalculateCostofEmail(df,igdf):
  cost = []
  for i in range(len(df)):
    costofemail=0
    for j in df.columns:
      if df.iloc[i,df.columns.get_loc(j)]==1:
        for k in range(len(igdf)):
          if j==igdf.iloc[k,0]:
            costofemail = costofemail+igdf.iloc[k,5]
    cost = np.append(cost,costofemail)
  return cost
```

Calculating Avg Cost Of Ham Emails and Spam Emails

```
def FindHamSpamAvgCost(df,labeldf):
  avghamcost = 0
  avgspamcost = 0
  lenham=0
  lenspam=0
  for i in range(len(df)):
    if labeldf.iloc[i,0]==1:
      avghamcost = avghamcost+df.iloc[i,df.columns.get_loc('cost')]
      lenham = lenham+1
    else:
      avgspamcost = avgspamcost+df.iloc[i,df.columns.get_loc('cost')]
      lenspam = lenspam+1
  avghamcost = avghamcost/lenham
  avgspamcost = avgspamcost/lenspam
  return avghamcost,avgspamcost
```

Building Add Words

```
def Addwords(df,avghamcost,igdf):
  for i in range(len(df)):
    cost=0
    for j in df.columns:
      if df.iloc[i,df.columns.get_loc('cost')] > avghamcost:
        if df.iloc[i,df.columns.get_loc(j)]==1:
          df.iloc[i,df.columns.get_loc(j)]==0
          newdf=pd.DataFrame(data=df.iloc[[i]],columns=df.columns)
          cost = CalculateCostofEmail(newdf,igdf)+cost
          df.iloc[i,df.columns.get_loc('cost')] = cost
  return df
```

Calculating Average Cost

```
def CalculateTotalAvgCost(df):
  totalcost = 0
  totaldata = len(df)
  for i in range(len(df)):
    totalcost = totalcost+df.iloc[i,df.columns.get_loc('cost')]
  totalcost = totalcost/totaldata
  return totalcost
```

Calculating False Positive and False Negative Rates

```
def CalculateFPFN(df1,df2):
  FP = 0
  FN = 0
  FAll=len(df1)
  for i in range(len(df1)):
    if df1.iloc[i,0]==0:
      if df2.iloc[i,0]==1:
        FP = FP+1
    if df1.iloc[i,0]==1:
      if df2.iloc[i,0]==0:
        FN=FN+1
  FP = FP/FAll
  FN = FN/FAll
  return FP,FN
```
